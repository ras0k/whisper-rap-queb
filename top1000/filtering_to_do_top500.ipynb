```python
import os
import shutil

# 1. Create the target folder
os.makedirs('whisper_dataset', exist_ok=True)

# 2. Perform a shallow clone of the repository to download it quickly
!git clone --depth 1 https://github.com/ras0k/whisper-rap-queb.git temp_repo

# 3. Define the source path where the txt files are located
txt_folder_path = "temp_repo/top1000/dataset 50-500/txt_files"

# 4. Extract only files from the txt folder into the flat 'whisper_dataset' directory
if os.path.exists(txt_folder_path):
    for filename in os.listdir(txt_folder_path):
        if filename.endswith('.txt'):
            source_file = os.path.join(txt_folder_path, filename)
            destination_file = os.path.join('whisper_dataset', filename)
            # Move the file
            shutil.move(source_file, destination_file)

# 5. Clean up the downloaded temporary repository
shutil.rmtree('temp_repo')

print("Success! All txt files are now located directly in the 'whisper_dataset' folder.")
```

```python
import os

dataset_dir = 'whisper_dataset'

total_files = 0
total_words = 0
total_chars = 0
word_counts = []

# Scan the directory and calculate metrics
for filename in os.listdir(dataset_dir):
    if filename.endswith('.txt'):
        total_files += 1
        filepath = os.path.join(dataset_dir, filename)

        with open(filepath, 'r', encoding='utf-8') as f:
            text = f.read()

        words = text.split()
        num_words = len(words)
        num_chars = len(text)

        total_words += num_words
        total_chars += num_chars
        word_counts.append(num_words)

# Print the formatted summary
if total_files > 0:
    avg_words = total_words / total_files
    min_words = min(word_counts)
    max_words = max(word_counts)

    print(f"📊 Dataset Summary for '{dataset_dir}'")
    print("-" * 40)
    print(f"Total Text Files: {total_files:,}")
    print(f"Total Words:      {total_words:,}")
    print(f"Total Characters: {total_chars:,}")
    print("-" * 40)
    print(f"Average Words per File: {avg_words:.1f}")
    print(f"Shortest File (words):  {min_words:,}")
    print(f"Longest File (words):   {max_words:,}")
    print("-" * 40)
else:
    print(f"No .txt files found in the '{dataset_dir}' directory.")
```

```python
import os
import re

dataset_dir = 'whisper_dataset'

for filename in os.listdir(dataset_dir):
    if filename.endswith('.txt'):
        filepath = os.path.join(dataset_dir, filename)

        # Read the file
        with open(filepath, 'r', encoding='utf-8') as f:
            text = f.read()

        # Rule 1: Replace curly apostrophe with straight apostrophe
        text = text.replace('’', "'")

        # Rule 2 & 3: Remove quotes, double single-quotes, guillemets, and parentheses
        chars_to_remove = ['"', "''", "<<", ">>", "«", "»", "(", ")"]
        for char in chars_to_remove:
            text = text.replace(char, "")

        # Rule 4: Fix space before comma
        text = text.replace(" ,", ",")

        # Rule 5: Replace two or more dots (... or ..) with nothing
        text = re.sub(r'\.{2,}', '', text)

        # Rule 6: Replace y'a / Y'a with y a / Y a
        # (Comment these two lines out if you want to exclusively do this manually)
        text = re.sub(r"\by'a\b", "y a", text)
        text = re.sub(r"\bY'a\b", "Y a", text)

        # Rule 7: Replace double or more spaces with a single space
        # Done last to easily catch any extra spaces left behind by the removed characters
        text = re.sub(r' +', ' ', text).strip()

        # Save the cleaned text back to the file
        with open(filepath, 'w', encoding='utf-8') as f:
            f.write(text)

print("Success! All text files have been cleaned and formatted.")
```
