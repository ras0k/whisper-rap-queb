```python
import os
import shutil

# 1. Create the target folder
os.makedirs('whisper_dataset', exist_ok=True)

# 2. Perform a shallow clone of the repository to download it quickly
!git clone --depth 1 https://github.com/ras0k/whisper-rap-queb.git temp_repo

# 3. Define the source paths where the files are located
base_path = "temp_repo/top1000/dataset 50-500"
folders_to_extract = ["audio_files", "txt_files"]

# 4. Extract all files from both folders into the flat 'whisper_dataset' directory
for folder in folders_to_extract:
    folder_path = os.path.join(base_path, folder)
    if os.path.exists(folder_path):
        for filename in os.listdir(folder_path):
            source_file = os.path.join(folder_path, filename)
            destination_file = os.path.join('whisper_dataset', filename)
            # Move the file
            shutil.move(source_file, destination_file)

# 5. Clean up the downloaded temporary repository
shutil.rmtree('temp_repo')

print("Success! All audio and txt files are now located directly in the 'whisper_dataset' folder.")
```

```python
import os

# Define your directories and audio format
txt_dir = 'whisper_dataset'
audio_dir = 'whisper_dataset' # Change this if your audio files are in a different folder
audio_ext = '.mp3'            # Change this to match your audio format (.mp3, .flac, etc.)

# Extract filenames without extensions using sets for fast comparison
txt_files = {os.path.splitext(f)[0] for f in os.listdir(txt_dir) if f.endswith('.txt')}
audio_files = {os.path.splitext(f)[0] for f in os.listdir(audio_dir) if f.endswith(audio_ext)}

# Find the differences
missing_audio = txt_files - audio_files
missing_txt = audio_files - txt_files

print(f"🔍 Cross-Referencing Check")
print("-" * 40)
print(f"Total Text Files:  {len(txt_files):,}")
print(f"Total Audio Files: {len(audio_files):,}")
print("-" * 40)

if not missing_audio and not missing_txt:
    print("✅ Perfect match! Every text file has a corresponding audio file and vice versa.")
else:
    if missing_audio:
        print(f"⚠️ Missing Audio Files ({len(missing_audio)}):")
        for name in list(missing_audio)[:5]:
            print(f"   - Missing {name}{audio_ext} (found transcript)")
        if len(missing_audio) > 5:
            print(f"   ... and {len(missing_audio) - 5} more.")

    if missing_txt:
        print(f"\n⚠️ Missing Text Files ({len(missing_txt)}):")
        for name in list(missing_txt)[:5]:
            print(f"   - Missing {name}.txt (found audio)")
        if len(missing_txt) > 5:
            print(f"   ... and {len(missing_txt) - 5} more.")
```

```python
import os

file_to_delete = "metadata.csv"

# Find the file
print(f"Searching for '{file_to_delete}'...")
find_command = f"find . -name {file_to_delete}"
result = os.popen(find_command).read().strip()

if result:
    print(f"Found '{file_to_delete}' at: {result}")
    # Delete the file
    delete_command = f"rm {result}"
    os.system(delete_command)
    print(f"'{file_to_delete}' deleted successfully from: {result}")

    # Verify deletion
    result_after_delete = os.popen(find_command).read().strip()
    if not result_after_delete:
        print(f"Verification: '{file_to_delete}' is no longer found.")
    else:
        print(f"Verification: '{file_to_delete}' still exists at: {result_after_delete}")
else:
    print(f"'{file_to_delete}' not found in the current directory or its subdirectories.")
```

```python
import os
import csv

# The folder where your audio and text files live
data_dir = "./whisper_dataset"

# Create and open the metadata.csv file
with open(os.path.join(data_dir, "metadata.csv"), "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["file_name", "transcription"])  # Hugging Face requires these headers

    # Get a sorted list of all files in the directory
    sorted_files = sorted(os.listdir(data_dir))

    # Loop through the alphabetically sorted files
    for filename in sorted_files:
        if filename.endswith(".mp3"):
            # Find the matching .txt file
            txt_filename = filename.replace(".mp3", ".txt")
            txt_path = os.path.join(data_dir, txt_filename)

            # If the text file exists, read the lyrics and write to the CSV
            if os.path.exists(txt_path):
                with open(txt_path, "r", encoding="utf-8") as f:
                    lyrics = f.read().strip()
                writer.writerow([filename, lyrics])

print("metadata.csv created successfully and sorted alphabetically!")
```

```python
import os
import shutil

file_to_copy = "metadata.csv"
destination_path = "."

# Find the file
print(f"Searching for '{file_to_copy}'...")
find_command = f"find . -name {file_to_copy}"
result = os.popen(find_command).read().strip()

if result:
    print(f"Found '{file_to_copy}' at: {result}")

    # Copy the file to the current directory
    try:
        shutil.copy(result, destination_path)
        print(f"'{file_to_copy}' copied successfully to '{destination_path}'.")

        # Verify the copy
        if os.path.exists(os.path.join(destination_path, file_to_copy)):
            print(f"Verification: '{file_to_copy}' now exists in the current directory.")
        else:
            print(f"Verification: Failed to confirm '{file_to_copy}' in the current directory.")

    except Exception as e:
        print(f"Error copying file: {e}")
else:
    print(f"'{file_to_copy}' not found in the current directory or its subdirectories.")
```
