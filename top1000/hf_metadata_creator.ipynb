```python
import os

file_to_delete = "metadata.csv"

# Find the file
print(f"Searching for '{file_to_delete}'...")
find_command = f"find . -name {file_to_delete}"
result = os.popen(find_command).read().strip()

if result:
    print(f"Found '{file_to_delete}' at: {result}")
    # Delete the file
    delete_command = f"rm {result}"
    os.system(delete_command)
    print(f"'{file_to_delete}' deleted successfully from: {result}")

    # Verify deletion
    result_after_delete = os.popen(find_command).read().strip()
    if not result_after_delete:
        print(f"Verification: '{file_to_delete}' is no longer found.")
    else:
        print(f"Verification: '{file_to_delete}' still exists at: {result_after_delete}")
else:
    print(f"'{file_to_delete}' not found in the current directory or its subdirectories.")
```

```python
import os
import csv

# The folder where your audio and text files live
data_dir = "./whisper_dataset"

# Create and open the metadata.csv file
with open(os.path.join(data_dir, "metadata.csv"), "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["file_name", "transcription"])  # Hugging Face requires these headers

    # Loop through all files in the directory
    for filename in os.listdir(data_dir):
        if filename.endswith(".mp3"):
            # Find the matching .txt file
            txt_filename = filename.replace(".mp3", ".txt")
            txt_path = os.path.join(data_dir, txt_filename)

            # If the text file exists, read the lyrics and write to the CSV
            if os.path.exists(txt_path):
                with open(txt_path, "r", encoding="utf-8") as f:
                    lyrics = f.read().strip()
                writer.writerow([filename, lyrics])

print("metadata.csv created successfully!")
```

```python
import os
import shutil

file_to_copy = "metadata.csv"
destination_path = "."

# Find the file
print(f"Searching for '{file_to_copy}'...")
find_command = f"find . -name {file_to_copy}"
result = os.popen(find_command).read().strip()

if result:
    print(f"Found '{file_to_copy}' at: {result}")

    # Copy the file to the current directory
    try:
        shutil.copy(result, destination_path)
        print(f"'{file_to_copy}' copied successfully to '{destination_path}'.")

        # Verify the copy
        if os.path.exists(os.path.join(destination_path, file_to_copy)):
            print(f"Verification: '{file_to_copy}' now exists in the current directory.")
        else:
            print(f"Verification: Failed to confirm '{file_to_copy}' in the current directory.")

    except Exception as e:
        print(f"Error copying file: {e}")
else:
    print(f"'{file_to_copy}' not found in the current directory or its subdirectories.")
```
