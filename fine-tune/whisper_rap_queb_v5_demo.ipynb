```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("automatic-speech-recognition", model="ras0k/whisper-rap-queb-v5")
```

```python
# Load model directly
from transformers import AutoProcessor, AutoModelForSpeechSeq2Seq

processor = AutoProcessor.from_pretrained("ras0k/whisper-rap-queb-v5")
model = AutoModelForSpeechSeq2Seq.from_pretrained("ras0k/whisper-rap-queb-v5")
```

```python
import gradio as gr
from transformers import pipeline
import torch

# 1. Check for GPU
device = "cuda:0" if torch.cuda.is_available() else "cpu"
print(f"Loading model on: {device}")

# 2. Initialize the pipeline WITH chunking enabled
pipe = pipeline(
    "automatic-speech-recognition",
    model="ras0k/whisper-rap-queb-v5",
    device=device,
    chunk_length_s=30,  # <-- This slices long audio into 30s chunks automatically
    batch_size=8,       # <-- Speeds up processing by running chunks in parallel on the GPU
)

# 3. Define the transcription function
def transcribe(audio_path):
    if audio_path is None:
        return "Please upload or record an audio file."

    # We explicitly tell it to return text (and optionally timestamps)
    prediction = pipe(audio_path, return_timestamps=False)
    return prediction["text"]

# 4. Build and launch the Gradio Interface
demo = gr.Interface(
    fn=transcribe,
    inputs=gr.Audio(sources=["microphone", "upload"], type="filepath", label="Audio Input"),
    outputs=gr.Textbox(label="Transcription"),
    title="🎙️ Whisper Rap Queb v5",
    description="Upload any length of audio! The pipeline will automatically chunk and process files longer than 30 seconds."
)

demo.launch(debug=True)
```
